```python
class Kost:
    def __init__(self, nama, alamat, harga_per_bulan):
        self.nama = nama 
        self.alamat = alamat
        self.harga_per_bulan = harga_per_bulan
        self.penyewa = []

    def add_penyewa(self, penyewa):
        self.penyewa.append(penyewa)

    def list_penyewa(self):
        if not self.penyewa:
            return ["Belum ada penyewa"]
        else:
            return [penyewa.nama for penyewa in self.penyewa]

    def describe_kost(self):
        # Menampilkan deskripsi kost
        print(f"\nHi, selamat datang di {self.nama} :)\n")
        print("----------Deskripsi Kost----------\n")
        print(f"Nama Kost       : {self.nama}")
        print(f"Alamat          : {self.alamat}")
        print(f"Harga per Bulan : Rp.{self.harga_per_bulan}")

class KostPutra(Kost):
    def __init__(self, nama, alamat, harga_per_bulan, fasilitas=["AC", "WiFi", "Parkir"]):
        super().__init__(nama, alamat, harga_per_bulan)
        self.__fasilitas = fasilitas

    def show_fasilitas(self):
        return self.__fasilitas

    def describe_kost(self):
        # Menampilkan deskripsi kost putra dan fasilitasnya
        super().describe_kost()
        print("\nFasilitas Kost Putra:")
        for fasilitas in self.show_fasilitas():
            print(f"- {fasilitas}")

class KostPutri(Kost):
    def __init__(self, nama, alamat, harga_per_bulan, keamanan=["Satpam", "CCTV", "Akses Fingerprint"]):
        super().__init__(nama, alamat, harga_per_bulan)
        self.__keamanan = keamanan

    def show_keamanan(self):
        return self.__keamanan

    def describe_kost(self):
        # Menampilkan deskripsi kost putri dan keamanannya
        super().describe_kost()
        print("\nKeamanan Kost Putri:")
        for keamanan in self.show_keamanan():
            print(f"- {keamanan}")

class KostCampur(Kost):
    def __init__(self, nama, alamat, harga_per_bulan, kebijakan=["Tamu tidak boleh menginap", "Tamu maksimal berkunjung sampai jam 9 malam"]):
        super().__init__(nama, alamat, harga_per_bulan)
        self.__kebijakan = kebijakan

    def show_kebijakan(self):
        return self.__kebijakan

    def describe_kost(self):
        # Menampilkan deskripsi kost campur dan kebijakan yang berlaku
        super().describe_kost()
        print("\nKebijakan Kost Campur:")
        for kebijakan in self.show_kebijakan():
            print(f"- {kebijakan}")

class Penyewa:
    def __init__(self, nama, usia, jenis_kelamin):
        self.nama = nama
        self.usia = usia
        self.jenis_kelamin = jenis_kelamin

class KontrakSewa:
    def __init__(self, penyewa, kost, durasi_bulan):
        self.penyewa = penyewa
        self.kost = kost
        self.durasi_bulan = durasi_bulan 
        self.kost.add_penyewa(penyewa)
        self.total_biaya = self.kost.harga_per_bulan * self.durasi_bulan

    def describe_kontrak(self):
        # Menampilkan informasi kontrak sewa
        print("\n----------Kontrak Sewa----------\n")
        print(f"Nama Penyewa   : {self.penyewa.nama}")
        print(f"Nama Kost      : {self.kost.nama}")
        print(f"Durasi Sewa    : {self.durasi_bulan} bulan")
        print(f"Total Biaya    : Rp.{self.total_biaya}")

class Pembayaran:
    def __init__(self, kontrak, jumlah):
        self.kontrak = kontrak
        self.jumlah = jumlah

    def make_payment(self):
        self.kontrak.describe_kontrak()
        print(f"\nPembayaran sebesar Rp.{self.jumlah} telah dilakukan.\n")
        print("=" * 50)

class KostManager:
    def __init__(self):
        self.kost_list = []

    def add_kost(self, kost):
        self.kost_list.append(kost)

    def display_all_kost(self):
        for kost in self.kost_list:
            kost.describe_kost()
            print("\nPenyewa:")
            for penyewa in kost.list_penyewa():
                print(f"- {penyewa}")
            print("\n" + "=" * 50)

kost_manager = KostManager() # Membuat objek class manager

kost_putra = KostPutra("Kost Putra Djaya", "Jl. Mawar No. 31", 1900000) # Membuat objek-objek Kost
kost_putri = KostPutri("Kost Putri Bahagia", "Jl. Melati No. 3", 2500000)
kost_campur = KostCampur("Kost Campur Harmoni", "Jl. Anggrek No.5", 1500000)

kost_manager.add_kost(kost_putra) # Menambahkan objek Kost ke KostManager
kost_manager.add_kost(kost_putri)
kost_manager.add_kost(kost_campur)

penyewa1 = Penyewa("Bayu Adjie ", 24, "Laki-laki") # Membuat objek Penyewa pertama
kontrak1 = KontrakSewa(penyewa1, kost_putra, 6) # Membuat objek KontrakSewa untuk penyewa pertama, menyewa Kost Putra selama 6 bulan
pembayaran1 = Pembayaran(kontrak1, 11400000) # Membuat objek Pembayaran untuk kontrak pertama, dengan total pembayaran sebesar 11.400.000
pembayaran1.make_payment() # Melakukan pembayaran untuk kontrak pertama

penyewa2 = Penyewa("Elisa Dwi", 22, "Perempuan") # Membuat objek Penyewa kedua
kontrak2 = KontrakSewa(penyewa2, kost_putri, 12) # Membuat objek KontrakSewa untuk penyewa kedua, menyewa Kost Putri selama 12 bulan
pembayaran2 = Pembayaran(kontrak2, 30000000) # Membuat objek Pembayaran untuk kontrak kedua, dengan total pembayaran sebesar 30.000.000
pembayaran2.make_payment() # Melakukan pembayaran untuk kontrak kedua

kost_manager.display_all_kost() # Menampilkan semua kost yang dikelola oleh KostManager
```
